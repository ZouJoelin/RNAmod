```python
# merge fast5.csv & fastq.csv to creat dataset pairs...
import pandas as pd

fast5_pd = pd.read_csv("./fast5/unmod_fast5_list.csv")
fastq_pd = pd.read_csv("./fastq/unmod_fastq_list.csv")

# print(fast5_pd)
# print(fastq_pd)

merged_df = fast5_pd.merge(fastq_pd, on=['read','channel'], how='right')
merged_df
```

```python
from data_utils import correct_by_reference, seq2label

reference = "./reference.fasta"
merged_df["fastq_seq"] = correct_by_reference(merged_df["fastq_seq"], reference)

modify = "m6A"
merged_df["label"] = seq2label(merged_df["fastq_seq"], modify)
```

```python

merged_df.to_csv("./fast5q_label.csv", index=False)
merged_df
```
